```python
# Списки из задачи
my_list_1 = [2, 5, 8, 2, 12, 12, 4, 3]
my_list_2 = [2, 7, 12, 3]
# Создаем новый пустой список, поскольку большинство методов удаляют по индексу, либо .remove - не удаляет дубликаты, а .set смешивает индексы
my_list_new = []
# Создаем цикл for для наполнения пустого списка итерируя my_list_1 сравнивая с my_list_2
for i in my_list_1:
    if i not in my_list_2:
        my_list_new.append(i)
print(my_list_new)
```
