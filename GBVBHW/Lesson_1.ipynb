#### 4

```python
number = int(input("Введите целое положительное число: "))
max_digit = 0
while number > 0:
    digit = number % 10
    if digit > max_digit:
        max_digit = digit
    number = number // 10
print(f"Самая большая цифра в числе: {max_digit}")
```

#### 5-6

```python
proceed = int(input("Введите сумму выручки: "))
cost = int(input("Введите сумму расходов: "))
if proceed > cost:
    profit = (proceed / cost - 1) * 100
    profit = round(profit)
    number = int(input("Введите численность сотрудников: "))
    proceed_number = proceed / number
    proceed_number = round(proceed_number)
    print(f"Прибыль {proceed} — выручка больше издержек, \nРентабельность составила {profit}%, \nПрибыль на сотрудника составила {proceed_number}")
else:
    print(f"Убыток {cost} — издержки больше выручки.")
```

```python
a = int(input('Result first day '))
b = int(input('Final result  '))

Numder_Day = 1
result = a
print(f"Day number: {Numder_Day}: {result:.2f}")
while result < b:
    result *= 1.1
    Numder_Day += 1
    print(f"Day number: {Numder_Day}: {result:.2f}")
print(f"Answer: on {Numder_Day} Day runner will achieve the result")



```
