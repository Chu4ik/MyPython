```python
total_second = int(input('Введите количество секунд: '))
hour = total_second // 3600
minute = (total_second % 3600) // 60
seconds =  total_second % 60
print(hour, ':', minute, ':', seconds)
```

```python
first_count = int(input('Введите значение: '))
second_count = first_count + (first_count * 10)
third_count = first_count * 100 + second_count      
sum = first_count + second_count + third_count
print(sum)
print(first_count, '+', second_count, '+', third_count)
```

```python
number = int(input("Введите целое положительное число: "))
max_digit = 0
while number > 0:
    digit = number % 10
    if digit > max_digit:
        max_digit = digit
    number = number // 10
print(f"Самая большая цифра в числе: {max_digit}")
```
