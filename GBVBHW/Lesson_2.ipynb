##### 2 ) Для списка реализовать обмен значений соседних элементов. Значениями обмениваются элементы с индексами 0 и 1, 2 и 3 и т. д. При нечётном количестве элементов последний сохранить на своём месте. Для заполнения списка элементов нужно использовать функцию input().

```python
# Вводим элементы списка через input(), разделенные пробелами
user_input = input("Введите элементы списка, разделенные пробелами: ")

# Преобразуем введенную строку в список
elements = user_input.split()

# Перебираем элементы списка с шагом 2
for i in range(0, len(elements) - 1, 2):
    # Обмениваем элементы местами
    elements[i], elements[i + 1] = elements[i + 1], elements[i]

# Выводим измененный список
print("Результат обмена соседних элементов:", elements)


```

##### 3 ) Пользователь вводит месяц в виде целого числа от 1 до 12. Сообщить, к какому времени года относится месяц (зима, весна, лето, осень). Напишите решения через list и dict.

```python
seasons = {
'winter': [12,1,2],
'spring': [3,4,5],
'summer': [6,7,8],
'autumn': [9,10,11]
}
# Запрашиваем месяц у пользователя
month = int(input("Введите номер месяца от 1 до 12: "))
# Итерируем по ключам и значениям словаря
for key, value_list in seasons.items():
    # Итерируем по элементам списка, который является значением для текущего ключа
    for item in value_list:
        if item == month:
            print(key)
```

##### 4 ) Пользователь вводит строку из нескольких слов, разделённых пробелами. Вывести каждое слово с новой строки. Строки нужно пронумеровать. Если слово длинное, выводить только первые 10 букв в слове.

```python
# Вводим элементы списка через input(), разделенные пробелами
user_input = input('Введите строку из нескольких слов, разделённых пробелами')
# Преобразуем введенную строку в список
elements = user_input.split()
# ограничиваем вывод 10 символами
max_len = 10
# перебираем список и присваеваем номер строки (индекс)
for index, line in enumerate(elements, start=1): # start=1 означает, что нумерация начнется с 1
    print(f"{index}) {line:.{max_len}}")
```

##### 5 ) Реализовать структуру «Рейтинг», представляющую собой набор натуральных чисел, который не возрастает. У пользователя нужно запрашивать новый элемент рейтинга. Если в рейтинге существуют элементы с одинаковыми значениями, то новый элемент с тем же значением должен разместиться после них.
Подсказка. Например, набор натуральных чисел: 7, 5, 3, 3, 2.
Пользователь ввёл число 3. Результат: 7, 5, 3, 3, 3, 2.
Пользователь ввёл число 8. Результат: 8, 7, 5, 3, 3, 2.
Пользователь ввёл число 1. Результат: 7, 5, 3, 3, 2, 1.
Набор натуральных чисел можно задать сразу в коде, например, my_list = [7, 5, 3, 3, 2].

```python
# базовый список
my_list = [7,5,3,3,2]
#Запрашиваем новые элементы рейтинга
user_input = int(input('Введите любое натуральное число'))
# создаем пустой список
new_list = my_list.copy()

# Проверяем, является ли введённое значение числом и больше ли оно нуля
while True:
    if user_input > 0:
        # итерируем список и добавляем элемент с условием
        for i in new_list:
            if i < user_input:
                new_list = [user_input] + new_list
            elif i == user_input:
                new_list.insert(i+1, user_input)
            else:
                new_list.append(user_input)
        # Выводим новый список
        print(f"Выводим новый список {new_list}")
        break
    else:
        print("Неправильный ввод. Пожалуйста, введите натуральное число.")



```
